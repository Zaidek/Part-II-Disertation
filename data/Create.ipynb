```python
# import required modules
import numpy as np
import pandas as pd
import datetime
import collections
from collections import Counter
import nltk
from nltk.corpus import stopwords
import pickle
```

```python
# import required pytoch modules
import torch
```

```python
# import data
data = pd.read_csv("~/Downloads/archive/hacker_news_sample.csv")
```

```python
# only collect stories
data = data.loc[data.type == "story"]
```

```python
# split data into train, validation, and test sets
seed = 123
distributions = [0.75, 0.1, 0.15]
split_data = torch.utils.data.random_split(dataset=data, lengths=distributions, generator=torch.Generator().manual_seed(seed))
train = data.iloc[split_data[0].indices]
valid = data.iloc[split_data[1].indices]
test = data.iloc[split_data[2].indices]
```

```python
# check lens are consistent
print(len(train), len(valid), len(test))
print(len(train) / len(data), len(valid) / len(data), len(test) / len(data))
```

```python
# save datasets
with open("../data/data_train", "wb") as fb:
    pickle.dump(train, fb)

with open("../data/data_valid", "wb") as fb:
    pickle.dump(valid, fb)

with open("../data/data_test", "wb") as fb:
    pickle.dump(test, fb)
```
